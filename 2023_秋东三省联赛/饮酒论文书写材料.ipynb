```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
data=pd.read_excel('modeling.xlsx')
data
```

```python
data1=data[['是否饮酒','饮酒年数','是否饮用高度白酒','饮用高度白酒频率','平均每次高度白酒饮用量','是否饮用低度白酒',
           '饮用低度白酒频率','平均每次低度白酒饮用量','是否饮用啤酒','饮用啤酒频率','平均每次啤酒饮用量','是否饮用黄酒或糯米酒',
           '饮用黄酒或糯米酒频率','平均每次黄酒或糯米酒饮用量','是否饮用葡萄酒','饮用葡萄酒频率','平均每次葡萄酒饮用量']]
```

```python
data5=data1[data1.是否饮酒==2]
data6=data1[data1.是否饮酒==1]
data7=data1[data1.是否饮酒==3]
columns_to_fill=['是否饮酒','饮酒年数','是否饮用高度白酒','饮用高度白酒频率','平均每次高度白酒饮用量','是否饮用低度白酒',
           '饮用低度白酒频率','平均每次低度白酒饮用量','是否饮用啤酒','饮用啤酒频率','平均每次啤酒饮用量','是否饮用黄酒或糯米酒',
           '饮用黄酒或糯米酒频率','平均每次黄酒或糯米酒饮用量','是否饮用葡萄酒','饮用葡萄酒频率','平均每次葡萄酒饮用量']
data7[columns_to_fill] = data7[columns_to_fill].fillna(0)
data5[columns_to_fill] = data5[columns_to_fill].fillna(0)
```

```python
data1 = pd.concat([data5,data6,data7],axis=0,ignore_index=False)
```

```python
data2=data1.describe()
data2.to_excel('饮酒数据基本描述.xlsx')
```

```python
def missing_percentage(df):
    """This function takes a DataFrame(df) as input and returns two columns, total missing values and total missing values percentage"""
    total = df.isnull().sum().sort_values(ascending = False)
    percent = round(df.isnull().sum().sort_values(ascending = False)/len(df)*100,2)
    return pd.concat([total, percent], axis=1, keys=['Total','Percent'])
```

```python
data3=missing_percentage(data1)
data3.to_excel('饮酒数据缺失率.xlsx')
```
